```python
# Dataset

template_startup_df = pd.read_csv("unicorns.csv")
```

Iremos usar uma amostra de 50 empresas por ordem decrescente de arbitragem de valor, logo, os valores nulos não serão tratados, pois, levando em conta o tamanho da amostra e dos tipos de dados que estão nulos, cidade e site, são informações que podém ser obtidas atráves de uma pesquisa rápida, caso seja necessário.

```python
# Renomeando as colunas

template_startup_df.rename(columns={"Updated at":"lastUpdate", "Company":"company", "Crunchbase Url":"crunchbaseUrl", "Last Valuation (Billion $)":"lastValuation",
                  "Date Joined":"joinDate", "Year Joined	":"yearDate", "City":"city", "Country":"country", "Industry":"industry", "Investors":"investors", "Company Website":"companyWebsite"}, inplace= True)
```

Pegando amostra de 50 empresas pela arbitragem de valor mais alta

```python
# Ordenando valores por ordem decrescente para retirar a amostra de 50 empresas

startups = template_startup_df.sort_values(by=["lastValuation"], ascending=False)
startups = startups.nlargest(50, "lastValuation")

# Conferindo a quantidade de linhas

startups.shape
```

```python
# 50 Empresas com maior arbitragem de valor

startups
```

```python
# Quantidade de empresas por país

startups.groupby("country")["lastValuation"].count()
```

```python
# Gráfico
startups.groupby("country")["lastValuation"].count().plot.barh(title="Quantidade de Empresas por País", color='midnightblue')
plt.xlabel("Quantidade")
plt.ylabel("País");
```

```python
# Empresas com a maior arbitragem de valor por país

startups.groupby("country")["lastValuation"].max()
```

```python
# Gráfico

startups.groupby("country")["lastValuation"].max().plot.barh(title='Maiores Arbitragens de Valorores por País', color='darkorchid')
plt.xlabel('Valor (Bilhão de Dólares)')
plt.ylabel('País');
```

```python
# Quantidade de empresas por tipo de segmento

startups.groupby("industry")["lastValuation"].count()
```

```python
# Gráfico

startups.groupby("industry")["lastValuation"].count().plot.barh(title='Quantidade de Empresas por Tipo de Segmento', color='mediumturquoise')
plt.xlabel('Quantidade')
plt.ylabel('Tipo de Segmento');
```

```python
# Empresas com a maior arbitragem de valor por tipo de segmento

startups.groupby("industry")["lastValuation"].max()
```

```python
# Gráfico

startups.groupby("industry")["lastValuation"].max().plot.barh(title='Maiores Arbitragens de Valorores por Tipo de Segmento', color='steelblue')
plt.xlabel('Valor (Bilhão de Dólares)')
plt.ylabel('Tipo de Segmento');
```
